```python
import folium
import gpxpy
import gpxpy.gpx
import pandas as pd

```

```python
m = folium.Map(
    location=(47.624248, -122.317024), 
    zoom_start=10,
    tiles="CartoDB.Voyager")
files = ['lake_loop.csv', 'out.csv','back.csv']
colors = ['#0783f7', '#075ff7','#0743f7']
for x in range(3):
    with open(files[x], 'r') as gpx:
        df = pd.read_csv(gpx)
        folium.PolyLine(df, 
        color=colors[x],
        clickable= True).add_to(m)

m
```
